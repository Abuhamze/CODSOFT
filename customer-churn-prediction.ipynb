```python
data=pd.read_csv('/kaggle/input/bank-customer-churn-prediction/Churn_Modelling.csv')
data
```

```python
data.info()
```

```python
newdata=data.drop(columns=['RowNumber','CustomerId','Surname','Gender','Tenure','HasCrCard'])
newdata
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
newdata['Geography']= label_encoder.fit_transform(newdata['Geography'])

newdata
```

```python
from sklearn.model_selection import train_test_split
x=newdata.drop(columns='Exited',axis=1)
y=newdata['Exited']
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import  classification_report,confusion_matrix
```

```python
lr = LogisticRegression()
lr.fit(train_x ,train_y)
print("Model Score on Training data",lr.score(train_x ,train_y))
print("Model Score on Testing data",lr.score(test_x ,test_y))
y_pred = lr.predict(test_x)
print(classification_report(y_pred ,test_y))


confusion_matrix(test_y, y_pred)

```

```python
rf=RandomForestClassifier()
rf.fit(train_x ,train_y)
print("Model Score on Training data",rf.score(train_x ,train_y))
print("Model Score on Testing data",rf.score(test_x ,test_y))
print(rf.feature_importances_,rf.feature_names_in_)
y_pred = rf.predict(test_x)
print(classification_report(y_pred ,test_y))


confusion_matrix(test_y, y_pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier()
gb.fit(train_x ,train_y)
print("Model Score on Training data",gb.score(train_x ,train_y))
print("Model Score on Testing data",gb.score(test_x ,test_y))
print(gb.feature_importances_)
y_pred = gb.predict(test_x)
print(classification_report(y_pred ,test_y))


confusion_matrix(test_y, y_pred)
```
